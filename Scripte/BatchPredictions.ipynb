```python
def trainAndSaveModel(name,pathTrain,pathTest,coinSide,freeze,freezelayer='conv5'):
    Coin = cnn.ConvNet(train_path= pathTrain, test_path =pathTest
                   , model=name,  img_size=(224, 224, 3),mint=coinSide)
    if freeze:
        print('Freeze layer')
        Coin.freeze(freezelayer)
    Coin.set_optimizer()
    Coin.load_dataset(128,val_split=0.1,)
    Coin.set_callbacks(callback_list=[], set_default=True)
    Coin.base_model.summary()
    Coin.model.summary()
    Coin.train(name)
    
    results = []
    for batch_images, batch_labels in Coin.train_set:
        predictions = Coin.model. (batch_images)

        # Store the activations and labels in the results list
        for prediction, label in zip(predictions, batch_labels):
            results.append([prediction.tolist(), label])

    data_array = np.array(results)

    # Save as a NumPy binary file
    np.save(name + '_train_' + coinSide + '_' +'.npy', data_array)
    
    
    results = []
    for batch_images, batch_labels in Coin.val_set:
        predictions = Coin.model.predict_on_batch(batch_images)

        # Store the activations and labels in the results list
        for prediction, label in zip(predictions, batch_labels):
            results.append([prediction.tolist(), label])

    data_array = np.array(results)

    # Save as a NumPy binary file
    np.save(name + '_val_' +coinSide + '_' +'.npy', data_array)
    
    
    results = []
    for batch_images, batch_labels in Coin.test_set:
        predictions = Coin.model.predict_on_batch(batch_images)

        # Store the activations and labels in the results list
        for prediction, label in zip(predictions, batch_labels):
            results.append([prediction.tolist(), label])

    data_array = np.array(results)

    # Save as a NumPy binary file
    np.save(name + '_test_' + coinSide + '_' +'.npy', data_array)
```
